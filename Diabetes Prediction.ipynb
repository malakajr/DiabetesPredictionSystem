# Part C - Implementation

```python
#numpy use for making numpy arrays
#pandas use for create data frames tables
#Importing the libraries related to the project
#StandardScaler use to standardize the data to a common range
#traintestsplit use for split data into training data and test data
#svm Support Vector Machine


import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### a)Input Data

In here author use Support Vector Machine(SVM) algorithm to implement this project. A set of labeled training data is necessary for Support Vector Machine (SVM) classifiers to learn from. A matrix with each row denoting a sample and each column denoting a feature should be used as the input data. The label or class of the sample should be in the final column. Integers, which are typically used to represent numbers, should be used as the class labels. Before using the input data to train the SVM classifier, preprocessing and normalization should be performed as needed.

The dataset used in this project got by Kaggle free downloadable. The dataset is preprocessed when it checked. All the data in this dataset is numeric and There are 9 attributes in dataset which are,

- 	Pregnancies – Number of times Pregnant
- 	Glucose - Plasma glucose concentration 2 hours in an oral glucose tolerance test
- 	BloodPressure – Diastolic blood pressure (mm Hg)
- 	SkinThickness – Triceps skin fold thickness (mm)
- 	Insulin – 2- Hour’s serum insulin (mu U/ml)
- 	BMI – Body Mass Index (weight in kg/height in m) ^ 2)
- 	DiabetesPedigreeFunction – Diabetes Pedigree Function
- 	Age – Age (Years)
- 	Outcome – Class Variable
    * 0 – non-Diabetic
    * 1 - Diabetic


```python
#loading the diabetes dataset to a pandas dataframe
# in here read comma seperated values files into Dataframe

diabetes_dataset = pd.read_csv('diabetes.csv')
```

```python
#printing the first 5 rows of the dataset
#user can get the last rows data sets by adding tail to head location

diabetes_dataset.head()
```

### b)Architecture Diagram (Training model and Output)



```python
#number and columns of the dataset

diabetes_dataset.shape
```

```python
#getting the statistical measure of the data

diabetes_dataset.describe()
```

```python
#testing the data and get a idea of how many non diabetes and diabetes in the dataset
#0 represents the non diabetes
#1 represents the diabetes

diabetes_dataset['Outcome'].value_counts()
```

```python
#geting the mean values for both non diabetes and diabetes

diabetes_dataset.groupby('Outcome').mean()
```

```python
#seperating the data and labels
#X user droping a particular column name outcome in dataset
#if user droping a column axis must be equal to 1 
#if user droping a rown axis must be equal to 0
#Y is getting only outcome dataset

X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
#user can get the all data except outcome

print(X)
```

```python
#user can get all the outcome results in here

print(Y)
```

### Data Standardization 

standardize the data in a particular range that helps machine learning model to make better predictions

```python
#taking one instance of standard scaler function

scaler = StandardScaler()
```

```python
#fit the data into scaler

scaler.fit(X)
```

```python
#Fitting all inconsistent data with standard scaler function
#Based on the standardization transforming all the data to a common range

standardized_data = scaler.transform(X)
```

```python
#getting the standardized data 
#all the values are in 0 and 1 will help to get better prediction

print(standardized_data)
```

```python
#In here X and Y use for training the machine learning model
#X represent the data and Y represents the model

X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

### Train Test Split

```python
#X train represents train machine learning model with X_train data
#X test represents once the model is trained evaluate the model with X_test data
#Y train represents all the labels for X_train data
#Y test represents all the labels which 1 or 0 for x_test data
#train test split will give four outputs and tha ouputs stored as x_train,x_test,y_train,y_test
#Entire data set given as X and entire labels given as Y
#test_size=0.2 represents 20% data for test data
#Stratify Y means splitted data outcome with 0 and 1
#if user doesn't include the stratify all the diabetics cases go to X_train non diabetics go to X_test
#random_state index for particular splitting data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
#in here 768 total participants in dataset
#614 participants going to use as training data
#154 participants use as test data

print(X.shape, X_train.shape, X_test.shape)
```

### Training The Model

```python
#SVC represents Support Vector Classifier
#in here user using linear model

classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
#Traing data is X_train
#Labels for the X_train data is Y_train

classifier.fit(X_train, Y_train)
```

# Part D – Testing

The holdout method was chosen by the author from a variety of testing techniques, including K-fold cross validation, Bootstrap sampling, and others. The dataset is split into two training sets and one testing set as this method moves forward. The model accuracy is evaluated by comparing the expected results with the actual results on the testing set. Below is a summary of the training and testing set's accuracy levels

```python
#Accuracy score in the training data
#Predic the label for all X_train data and storing all these labels in X_train_prediction variable
#Comparing the prediction of model with the original labels that is Y_train

X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data: ', training_data_accuracy)
```

```python
#Accuracy score in the test data

X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data: ', test_data_accuracy)
```
